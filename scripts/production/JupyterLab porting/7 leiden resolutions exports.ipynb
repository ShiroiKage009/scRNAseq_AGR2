```python
import scanpy as sc
import pandas as pd
import os
```

```python
in_path = 'S:/data cache/code_in_out/agr2/agr2_init_files/Aline/final dataset calculations from QC intermediate file/'
```

```python
out_path = 'S:/data cache/code_in_out/agr2/script_7_out/'
```

```python
aline = sc.read_h5ad(in_path + 'antrum_epithelium_inclLGR5_20241114.h5ad')
init = sc.read_h5ad('S:/data cache/code_in_out/agr2/agr2_init_files/agr2_nneigh30_pcs75_dir/agr2_bcharmony_meteuclidean_neighbors.h5ad')
```

```python
str(aline)
```

```python
ad = init[aline.obs_names.tolist()]
```

```python
# Remove columns in obs starting with 'leiden_res_'
columns_to_remove = [col for col in ad.obs.columns if col.startswith('leiden_res_')]
ad.obs.drop(columns=columns_to_remove, inplace=True)
```

```python
str(ad)
```

```python
sc.pp.pca(ad, n_comps = 50)
sc.pp.neighbors(ad, n_neighbors = 30, use_rep = 'X_harmony')
```

```python
# Set output directory for results
output_dir = out_path
os.makedirs(output_dir, exist_ok=True)
adata = ad

# Range of Leiden resolutions
resolutions = [round(0.1 * i, 1) for i in range(1, 14)]

for res in resolutions:
    # Perform Leiden clustering
    sc.tl.leiden(adata, resolution=res, key_added=f"leiden_{res}", n_iterations=2)
    
    # Perform differential gene expression analysis
    sc.tl.rank_genes_groups(adata, groupby=f"leiden_{res}", method="wilcoxon")
    
    # Create a DataFrame for each cluster and save to an Excel file
    excel_path = os.path.join(output_dir, f"leiden_res_{res}.xlsx")
    writer = pd.ExcelWriter(excel_path, engine="openpyxl")
    
    # Loop through each cluster
    for cluster in adata.obs[f"leiden_{res}"].unique():
        # Get the differential expression results for this cluster
        cluster_data = pd.DataFrame({
            "gene": adata.uns["rank_genes_groups"]["names"][cluster],
            "scores": adata.uns["rank_genes_groups"]["scores"][cluster],
            "logfoldchanges": adata.uns["rank_genes_groups"]["logfoldchanges"][cluster],
            "pvals": adata.uns["rank_genes_groups"]["pvals"][cluster],
            "pvals_adj": adata.uns["rank_genes_groups"]["pvals_adj"][cluster],
        })
        
        # Write this cluster to its own sheet
        cluster_data.to_excel(writer, sheet_name=f"Cluster_{cluster}", index=False)
    
    # Save the Excel workbook
    writer.close()

print("Processing complete. Results saved in the 'leiden_results' directory.")
```

```python
# Select columns that start with 'leiden_'
leiden_columns = [col for col in adata.obs.columns if col.startswith('leiden_')]

# Create a new DataFrame with these columns
clustree_df = adata.obs[leiden_columns].copy()

# Replace '.' with '_' in the column names
clustree_df.columns = [col.replace('.', '_') for col in clustree_df.columns]

# Optionally, include the cell IDs by resetting the index
clustree_df.reset_index(inplace=True)

# Export to CSV without the index column
clustree_df.to_csv(out_path + 'all_leidens_20241115_clustree_prep.csv', index=False)
```
