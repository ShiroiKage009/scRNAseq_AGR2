```python
import scanpy as sc
import anndata as ad
import pandas as pd
```

```python
from pybiomart import Server
```

```python
sc.settings.verbosity = 3
fig_res = 300
sc.set_figure_params(dpi = fig_res, frameon = 0)
path = '/home/achmed/ssdmnt/cache/external_datasets/'
```

```python
# Connect to the BioMart server
server = Server(host='http://www.ensembl.org')
```

```python
# Get the dataset you want to work with
dataset = server.marts['ENSEMBL_MART_ENSEMBL'].datasets['hsapiens_gene_ensembl']

# Query the dataset for Ensembl Gene IDs and corresponding Gene Symbols
conversion_table = dataset.query(attributes=['ensembl_gene_id', 'external_gene_name'])
```

```python
# Show first few rows of the conversion table
print(conversion_table.head())
```

```python
sc.settings.verbosity = 3
fig_res = 150
sc.set_figure_params(dpi = fig_res, frameon = 0)
```

```python
adata = sc.read('/home/achmed/ssdmnt/cache/external_datasets/gim and eim unified umap of all file.h5ad')
```

```python
str(adata)
```

```python
adata.obs['tissue'].unique()
```

```python
print(conversion_table.columns)
```

```python
# Get the gene names (assumed to be Ensembl IDs)
ensembl_ids = adata.var_names

# Create a DataFrame from the conversion table
conversion_df = pd.DataFrame(conversion_table)

# Set the Ensembl IDs as the index of the conversion DataFrame
conversion_df.set_index('Gene stable ID', inplace=True)

# Map the Ensembl IDs in your AnnData object to Gene Symbols
gene_symbols = ensembl_ids.map(conversion_df['Gene name'])

# Replace the var_names in the AnnData object
adata.var_names = gene_symbols
```

```python
sc.pl.embedding(adata, basis = 'X_umap_MinDist_0.3_N_Neighbors_15', color = ['tissue'], use_raw = False)
```

```python
epithelial = ['stem cell', 
              'parietal cell', 
              'enteroendocrine cell', 
              'intestine goblet cell', 
              'peptic cell', 
              'type G enteroendocrine cell', 
              'enterocyte', 
              'mucous neck cell',
              'foveolar cell of stomach',
              'epithelial cell of esophagus',
              'P/D1 enteroendocrine cell',
              'glandular cell of esophagus'
             ]
stromal = ['fibroblast', 
           'mast cell', 
           'capillary endothelial cell',
           'endothelial cell of artery',
           'vein endothelial cell'
          ]
```

```python
adata.var.index = adata.var.index.astype(str)
adata.write_h5ad(path + 'external_geneid_unproc.h5ad')
```
